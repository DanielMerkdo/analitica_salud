```python
import pandas as pd
import joblib  # o puedes usar pickle
from sklearn.preprocessing import StandardScaler  # si tu modelo necesita escalar

# Paso 1: Cargar datos
df = pd.read_csv('data/Datos_despliegue.csv')

# Paso 2: Preprocesamiento (si tu modelo fue entrenado con datos escalados o transformados)
# Por ejemplo, escalar los datos si usaste StandardScaler durante el entrenamiento
#scaler = joblib.load('salidas/escalador.pkl')  # si usaste un scaler guardado
#X = scaler.transform(df)  # suponiendo que df ya tiene solo las columnas relevantes
X = df  # usa los datos tal como están si no se requiere escalar


# Paso 3: Cargar el modelo entrenado
modelo = joblib.load('salidas/best_rf_optuna.pkl')  # reemplaza con el nombre real

# Paso 4: Realizar predicciones
predicciones = modelo.predict(X)
probabilidades = modelo.predict_proba(X)[:, 1]  # si quieres también las probabilidades

# Paso 5: Guardar resultados
resultados = df.copy()
resultados['Prediccion'] = predicciones
resultados['Probabilidad'] = probabilidades

resultados.to_excel('salidas/resultados_despliegue.xlsx', index=False)
```
